```python
# Q2
from blockchain import blockexplorer
import collections

# 取出最新的block和block height
latest_block = blockexplorer.get_latest_block()
block_height = latest_block.height
block_inputAddress = []
block_outputAddress = []


for i in range(2):
    block = blockexplorer.get_block_height(block_height-i)[0]
    txs = block.transactions
    
# 第一筆coinbase沒有input address,故從第二筆交易開始加進input,output的address list
    for t in txs[1:]:
        inputs = t.inputs
        outputs = t.outputs
        for input in inputs:
            inputAddress = input.address
            block_inputAddress.append(inputAddress)
        for output in outputs:
            outputAddress = output.address
            block_outputAddress.append(outputAddress)

# 補加入coinbase的output address
    firstoutput = txs[0].outputs[0]
    outputAddress = firstoutput.address
    block_outputAddress.append(outputAddress)

# 利用collections 找出top address
addressCounter = collections.Counter(block_inputAddress+block_outputAddress)

print(len(block_inputAddress+block_outputAddress))
print(addressCounter.most_common(20))
        
```

```python
# Q3
from blockchain import blockexplorer
from collections import defaultdict

# 取出最新的block和block height
latest_block = blockexplorer.get_latest_block()
block_height = latest_block.height
block_output = defaultdict(list)
aggregate = 0
block_addressAggregate = {}

# 找出前25筆block coinbase的地址及其amount
for i in range(25):
    block = blockexplorer.get_block_height(block_height-i)[0]
    txs = block.transactions
    firstOutput = txs[0].outputs
    outputAddress = firstOutput[0].address
    outputValue = firstOutput[0].value
    block_output[outputAddress].append(outputValue)
    
# 計算coinbase相同address的vlaue總和 
    for k,v in block_output.items():
        aggregate = 0 
        for i in v:
            aggregate += i
            block_addressAggregate[k] = aggregate
            
print(block_output)
print(block_addressAggregate)
```

```python
# Q4
from blockchain import blockexplorer
import collections

# 取出最新的block和block height
latest_block = blockexplorer.get_latest_block()
block_height = latest_block.height
block_inputAddress = []
block_outputAddress = []
compare_message = []

# 將input,output的address分別存入不同的list,再對比較兩個address是否有相同的值
for i in range(25):
    block = blockexplorer.get_block_height(block_height-i)[0]
    txs = block.transactions
    for t in txs[1:]:
        inputs = t.inputs
        outputs = t.outputs
        for input in inputs:
            inputAddress = input.address
            block_inputAddress.append(inputAddress)
        for output in outputs:
            outputAddress = output.address
            block_outputAddress.append(outputAddress)
compare_message = set(block_inputAddress)&set(block_outputAddress)

#判斷input,output皆有的地址是否存在
if len(compare_message) == 0 :
    print("No such address")
else :
    print(set(compare_message))
```
