```python
import requests
import pandas as pd
import numpy as np
import pickle
from bs4 import BeautifulSoup


response = requests.get('https://technews.tw/')
soup = BeautifulSoup(response.text, 'html.parser')

category_list = []
title_list = []
url_list = []

for content in soup.find_all('li', class_='block2014'):    
    category = content.find('div', class_='cat01').text  
    rows = content.find_all('a', href = True)
    for i in range(len(rows)):
        title = rows[i].text.replace("\n", "")
        category_list.append(category)
        url = rows[i]['href']
        title_list.append(title)
        url_list.append(url)
        
index_list = list(range(1,len(title_list)+1))
data = {'分類': category_list,'標題': title_list,'網址':url_list}
df = pd.DataFrame(data, columns = ["分類", "標題", "網址"], index = index_list)
pd.to_pickle(df,"/Users/Tinatien/Desktop/df.pkl")
df



   
```
