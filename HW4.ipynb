```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
from sklearn.linear_model import LinearRegression

%matplotlib inline 
%pylab inline

df = pd.read_csv('/Users/Tinatien/Desktop/Kids198.csv')
```

```python
#身高與體重的關係
seaborn.jointplot(x="Height", y="Weight", data=df, kind='reg', fit_reg= True, size=7)
plt.show()
```

```python
# 年齡與體重的關係
seaborn.jointplot(x="Age", y="Weight", data=df, kind='reg', fit_reg=True, size=7)
plt.show()
```

```python
# 年齡與身高的關係
seaborn.jointplot(x="Age", y="Height", data=df, kind='reg', fit_reg=True, size=7)
plt.show()
```

```python
# 同年齡的孩童，性別與身高的關係
seaborn.lmplot(x="Age", y="Height", data=df, hue="Sex")
```

```python
# 同年齡的孩童，性別與體重的關係
seaborn.lmplot(x="Age", y="Weight", data=df, hue="Sex")
```

```python
# 同年齡的孩童，種族與體重的關係
seaborn.lmplot(x="Age", y="Weight", data=df, hue="Race")
```

```python
# 同年齡的孩童，種族與身高的關係
seaborn.lmplot(x="Age", y="Height", data=df, hue="Race")
```
